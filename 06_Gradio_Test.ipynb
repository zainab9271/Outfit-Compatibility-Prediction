# Import Libraries

```python
import gradio as gr
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
import numpy as np
import shutil
import os
```

# Test Data

```python
test_outfit_df = pd.read_csv('/content/drive/MyDrive/Outfit_Recommendation_Project/Normal_Outfits_Data/test4siamese.csv')
test_outfit_df
```

```python
example_df = pd.DataFrame(columns=['tops', 'bottoms', 'shoes', 'outerwear', 'bags', 'compatibility'])

for i in range(10):
    samp = test_outfit_df.sample(1)
    samp.reset_index(drop=True, inplace=True)
    new_paths = []
    for item_path in samp.iloc[0,:-1]:
        new_paths.append(os.path.basename(shutil.copy(item_path, '/content')))
    new_paths.append(samp.iloc[0,-1])
    example_df = example_df.append(dict(zip(list(example_df.columns), new_paths)), ignore_index=True)
    


example_df
```

```python
example_log = example_df.iloc[:,:-1]
example_log
```

```python
model = tf.keras.models.load_model('/content/drive/MyDrive/Outfit_Recommendation_Project/best_model_4_normal_outfits')
model.summary()
```

# Gradio for Our Model

```python
def predict_outfit_compatibility(image1, image2, image3, image4, image5):
    
    image1 = image1/255.0
    image1 = tf.expand_dims(image1, axis=0)

    image2 = image2/255.0
    image2 = tf.expand_dims(image2, axis=0)

    image3 = image3/255.0
    image3 = tf.expand_dims(image3, axis=0)

    if image4 is None:
        image4 = plt.imread('/content/drive/MyDrive/Outfit_Recommendation_Project/blank_image.jpg').astype('float32')
    else:
        image4 = image4/255.0
    image4 = tf.expand_dims(image4, axis=0)

    if image5 is None:
        image5 = plt.imread('/content/drive/MyDrive/Outfit_Recommendation_Project/blank_image.jpg').astype('float32')
    else:
        image5 = image5/255.0
    image5 = tf.expand_dims(image5, axis=0)

    prediction = model.predict((image1, image2, image3, image4, image5))
    return {'Compatible':float(prediction[0][0]), 'Incompatible':float(1 - prediction[0][0])}
```

```python
gr.Interface(
    fn=predict_outfit_compatibility, 
    inputs=[gr.inputs.Image(shape=(224,224), label='Top'),
            gr.inputs.Image(shape=(224,224), label='Bottom'),
            gr.inputs.Image(shape=(224,224), label='Shoes'),
            gr.inputs.Image(shape=(224,224), label='Outerwear', optional=True),
            gr.inputs.Image(shape=(224,224), label='Bags', optional=True)
            ],
    outputs = gr.outputs.Label(num_top_classes=2),
    title='Outfit Compatibility Inspector',
    description='Pick up at least 3 items and check if your selected items are compatible to be considered as an outfit.',
    theme='peach',
    examples=example_log.to_numpy().tolist()
    ).launch()
```
