# Remove Undesired Items from Outfit Samples

Remove Items of Categories:

* hats
* accessories
* jewellery
* sunglasses
* scarves







```python
def parse_sample(outfit_sample):
    '''
    outfit_sample: STRING
    returns list of items id
    '''
    outfit_sample = re.sub('[\[\]\s\']', '', outfit_sample)
    outfit_items = np.array(outfit_sample.split(','))
    return outfit_items
```

```python
def parse_outfit(outfit_items, train_test_valid = 'train' ):
    '''
    outfit_items: list of outfit items
    returns dictionary of (outfit_item, item_path)
    '''
    if train_test_valid == 'train':
        outfit_imgs_path = train_outfits_imgs_df.query(f'outfits_images_id in {outfit_items.tolist()}')
    elif train_test_valid == 'test':
        outfit_imgs_path = test_outfits_imgs_df.query(f'outfits_images_id in {outfit_items.tolist()}')
    else:
        outfit_imgs_path = valid_outfits_imgs_df.query(f'outfits_images_id in {outfit_items.tolist()}')
    
    outfit_imgs_path.set_index('outfits_images_id', drop=True, inplace=True)
    
    return outfit_imgs_path.to_dict()['outfits_images_path']
    
```

```python
def get_item_id(outfit_items_path):
    '''
    outfit_items_path: dictionary of (outfit_item, item_path)
    returns dictionary of (outfit_item, item_id)
    '''
    outfit_items_id = {}
    for key, value in outfit_items_path.items():
        outfit_items_id[key] = item_img_df[item_img_df.item_image_path == value].item_id.to_numpy()[0]

    return outfit_items_id
```

```python
def get_items_category(outfit_items_id):
    '''
    outfit_items_id: dictionary of (outfit_item, item_id)
    returns dictionary of (outfit_item, category)
    '''
    outfit_items_category = {}
    for key, value in outfit_items_id.items():
        try:
            outfit_items_category[key] = item_metadata_df.query(f'item_id == {value}').semantic_category.to_numpy()[0]
        except:
            outfit_items_category[key] = 'NON_AVAILABLE'
    
    return outfit_items_category
```

```python
def remove_undesired_items(outfit_sample, train_test_valid = 'train'):
    '''
    outfit_sample: string
    '''
    undesired_categories = ('hats','accessories','jewellery','sunglasses','scarves', 'NON_AVAILABLE')
    outfit_items = parse_sample(outfit_sample)
    
    outfit_items_path = parse_outfit(outfit_items, train_test_valid)
    
    outfit_items_id = get_item_id(outfit_items_path)
    
    outfit_items_category = get_items_category(outfit_items_id)

    #in case we have several items of the same category, we only take one item from each category
    group_items_by_categ = {}
    for key, value in outfit_items_category.items():
        if not group_items_by_categ.get(value):
            group_items_by_categ[value] = []
        group_items_by_categ[value].append(key)
    
    for category, outfit_items in group_items_by_categ.items():
        group_items_by_categ[category] = outfit_items[np.random.randint(len(outfit_items))]

    # modified_sample = np.array([])
    modified_sample = {'tops':0, 'bottoms':0, 'shoes':0, 'outerwear':0, 'bags':0, 'all-body':0}
    for category, outfit_item in group_items_by_categ.items():
        if category not in undesired_categories:
            modified_sample[category] = outfit_item

    modified_sample = np.array2string(np.array(list(modified_sample.values())), separator=',')

    return modified_sample
```

What we will consider as outfits:
1.   Normal Outfits:

> * Tops
> * Bottoms
> * Shoes
> * Outerwear (optional)
> * Bags (optional)

2.   All-Body Outfits:

> * All-body
> * Shoes
> * Outerwear (optional)
> * Bags (optional)


Prepare Normal Outfits Data and All-body Outfits Data

```python
normal_outfits_df = pd.DataFrame(columns=['compatibility', 'outfit_sample'])
allBody_outfits_df = pd.DataFrame(columns=['compatibility', 'outfit_sample'])
```

```python
def process_outfit(outfit_sample, compatibility):
    outfit_items = parse_sample(outfit_sample)
    #index 0 = 'tops', 1 = 'bottoms', 2 = 'shoes', 3 = 'outerwear', 4 = 'bags', 5 = 'all-body'
    global normal_outfits_df
    global allBody_outfits_df
    if outfit_items[2] != '0':
        if outfit_items[0] != '0' and outfit_items[1] != '0':
            normal_outfits_df = normal_outfits_df.append({'compatibility':compatibility,
                                                          'outfit_sample':np.array2string(
                                                              np.array(outfit_items[:-1]), 
                                                              separator=',')
                                                          }, 
                                                         ignore_index=True)
        elif outfit_items[5] != '0':
            allBody_outfits_df = allBody_outfits_df.append({'compatibility':compatibility,
                                                            'outfit_sample':np.array2string(
                                                                np.array(outfit_items[-1:-5:-1]),
                                                                separator=',')
                                                            },
                                                           ignore_index=True) 

    
```

## Process Train Outfits

```python
for i in tqdm.tqdm(np.arange(train_df.shape[0])):
    compatibility, outfit_sample = train_df.iloc[i]
    process_outfit(outfit_sample, compatibility)
```
